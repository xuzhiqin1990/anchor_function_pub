```python
import numpy as np
working_dir_LSTM=r'/root/code/LLM/anchor-function-new/result/GPT_anchorlength_normal'
acc_test_unmask_his_mean_all=[]
acc_test_unmask_his_std_all=[]
lsp_list=['0.0', '0.05', '0.1', '0.15', '0.2', '0.25', '0.3', '0.35', '0.4', '0.45', '0.5', '0.55', '0.6', '0.65', '0.7', '0.75', '0.8', '0.85', '0.9', '0.95', '1.0']
acc_dict=np.load('/root/code/LLM/plot_for_anchor_func/anchor_length_task_acc_from_jiaweile.npz', allow_pickle=True)



test_3_lst=[]
test_23_lst=[]
test_123_lst=[]

# for seed in ['1', '2', '3']:
#     for N in ['250', '300']:


for seed in ['1']:
    for N in ['300']:
        for lsp in lsp_list[::2]:
        # for lsp in ['1.0']:
            acc_list=acc_dict[lsp].item()[seed][N]
            if acc_list!=[]:
                # acc_list_max=[max(i) for i in acc_list]
                # print(acc_list_max)
                print(lsp, seed, N)
                test_3_lst.append(max(acc_list[0]))
                test_23_lst.append(max(acc_list[1]))
                test_123_lst.append(max(acc_list[2]))



for seed in ['2']:
    for N in ['300']:
        for lsp in ['1.0']:
            acc_list=acc_dict[lsp].item()[seed][N]
            if acc_list!=[]:
                # acc_list_max=[max(i) for i in acc_list]
                # print(acc_list_max)
                # print(lsp, seed, N)
                test_3_lst.append(max(acc_list[0]))
                test_23_lst.append(max(acc_list[1]))
                test_123_lst.append(max(acc_list[2]))











```
